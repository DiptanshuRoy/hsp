```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
data_path = Path("../data/processed/cleaned_data.parquet")
df=pd.read_parquet(data_path)
```

```python
df.head()
```

```python
med_sp=pd.get_dummies(df,columns=['medical_specialty'],drop_first=True,dtype='int')
```

```python
med_cols = [
    'metformin', 'repaglinide', 'nateglinide', 'chlorpropamide',
    'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide',
    'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
    'tolazamide', 'examide', 'citoglipton', 'insulin',
    'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone'
]

med_charged = pd.get_dummies(df, columns=med_cols, drop_first=True, dtype='int')
```

```python
race=pd.get_dummies(df,columns=['race'],drop_first=True,dtype='int')
gen=pd.get_dummies(df,columns=['gender'],drop_first=True,dtype='int')
age=pd.get_dummies(df,columns=['age'],drop_first=True,dtype='int')
```

```python
# Convert '>30' and '<30' to 1 (readmitted), 'NO' to 0
df['readmitted'] = df['readmitted'].apply(lambda x: 1 if x in ['<30', '>30'] else 0)
```

```python
# 1. See duplicated column names
print(df.columns[df.columns.duplicated()])
```

```python
# 2. Drop duplicated columns
df = df.loc[:, ~df.columns.duplicated()]
```

```python
print(type(df['number_outpatient']))  # should now be <class 'pandas.core.series.Series'>
```

```python
df['visit_total'] = (
    df['number_outpatient'] +
    df['number_emergency'] +
    df['number_inpatient']  # ✅ This should be a Series, not a list!
)
```

```python


# Fix column names


# Make sure column name is correct (you wrote `number_impatient`, it should be `number_inpatient`)
df['visit_ratio'] = df['number_emergency'] / (df['number_inpatient'] + 1)

# Also, you wrote 'number_medications' as 'number_medications' in the column
df['med_per_day'] = df['num_medications'] / (df['time_in_hospital'] + 1)

# Correct log transformation
df['log_num_med'] = np.log1p(df['num_medications'])

# Optional: log of total visits
df['log_total_visits'] = np.log1p(1+df['visit_total'])
df['log_num_med']=np.log1p(1+df['num_medications'])



```

```python
df.head()
```

```python
import pandas as pd

# Medication Binning
df["med_bin"] = pd.cut(
    df["num_medications"],
    bins=[0, 10, 20, 40, 100],
    labels=["low", "medium", "high", "very_high"],
    right=False  # intervals will be like [0,10), [10,20) ...
)

# Diagnoses Binning
df["diag_bin"] = pd.cut(
    df["number_diagnoses"],
    bins=[0, 4, 8, 12, 20],
    labels=["low", "medium", "high", "very_high"],
    right=False
)

# Time in Hospital Binning
df["stay_bin"] = pd.cut(
    df["time_in_hospital"],
    bins=[0, 3, 7, 15],
    labels=["short", "medium", "long"],
    right=False
)
```

```python
df.head()
```

```python
# Create dummies for all three binned columns
binned_dummies = pd.get_dummies(df[["med_bin", "diag_bin", "stay_bin"]], prefix=["med", "diag", "stay"],dtype='int')

# Concatenate with original df
df = pd.concat([df, binned_dummies], axis=1)

# Optional: drop the original string columns
df.drop(["med_bin", "diag_bin", "stay_bin"], axis=1, inplace=True)
```
